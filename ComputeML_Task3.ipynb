```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
import seaborn as sns
```

```python
df=pd.read_csv("haberman (1).csv")
df.columns=['patient_age','operation_year','positive_auxillary_nodes','survival_status']
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
for i in range(len(df)):
    if(df['survival_status'][i]==1):
        df['survival_status'][i]=0
    else:
        df['survival_status'][i]=1
```

```python
print(df.head())
print(df.tail())
```

```python
corr=df.corr()
sns.heatmap(corr, annot = True, linewidth = 1, cbar = True, center = 1)
```

```python
##1.is the age group and positive axillary nodes correlated?
```

```python
plt.figure(figsize=(25,15))
sns.lineplot(x='patient_age',y='positive_auxillary_nodes',hue='survival_status',data=df)
```

```python
#hence we can say that yes positive auxillary nodes and patient age can be said to have positive correlation. 
#4.from the heatmap we can observe axillary nodes have a high effect on survival satus.
#more is the no. of auxillaries more is the survival rate.
```

```python
##2.does a specific age group have more cances of surviving 
```

```python
plt.figure(figsize=(25,15))
sns.histplot(data = df, x = "patient_age", hue = "survival_status")
```

```python
#hence we can say that there is more chance of survival in age range 35-40 and 55-60
```

```python
#3.does the operation year have any effect on the survival status
```

```python
plt.figure(figsize=(25,15))
sns.countplot(x='operation_year',hue='survival_status',data=df)
```

```python
plt.figure(figsize=(25,15))
sns.lineplot(x='operation_year',y='survival_status',data=df)
```

```python
# hence we can say that year 61 had highest survival rate, whereas year 65 had lowest survival rate.
# more the survival status is tendig towards 0 more is the rate of survival.
```
