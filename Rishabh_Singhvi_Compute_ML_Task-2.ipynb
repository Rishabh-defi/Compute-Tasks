```python
####Replace NaN values with average of that column

#Initial array :  

#[[ 1.3  2.5  3.6  nan]
# [ 2.6  3.3  nan  5.5]
# [ 2.1  3.2  5.4  6.5]]
```

```python
old_array = np.array([[1.3, 2.5, 3.6, np.nan], 
                      [2.6, 3.3, np.nan, 5.5],
                      [2.1, 3.2, 5.4, 6.5]])
  
print ("old array", old_array)
  
column_mean = np.nanmean(old_array, axis = 0)
  
print ("columns mean", str(column_mean))
  
inds = np.where(np.isnan(old_array))
  
old_array[inds] = np.take(column_mean, inds[1])

print ("new array", old_array)
```
